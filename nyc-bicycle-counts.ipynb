```
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```
df = pd.read_csv("nyc-bicycle-counts.csv")
```

```
df.dtypes
```

```
df.head()
```

```
# Number of rows
df.shape[0]
```

```
g = sns.pairplot(df, vars = ["High Temp (¬∞F)","Low Temp (¬∞F)","Precipitation","Total"])
```

```
fig,ax=plt.subplots(figsize=(10,8))
sns.boxplot(data=df[['High Temp (¬∞F)','Low Temp (¬∞F)']])
ax.set_title('HighTemp_LowTemp_outliers')
plt.show()
```

```
fig,ax=plt.subplots(figsize=(10,8))
sns.boxplot(data=df[['Precipitation']])
ax.set_title('Precipitation_outliers')
plt.show()
```

```
fig,ax=plt.subplots(figsize=(10,8))
sns.boxplot(data=df[['Brooklyn Bridge','Manhattan Bridge','Williamsburg Bridge','Queensboro Bridge','Total']])
ax.set_title('count_outliers')
plt.show()
```

```
df['Day.1']=df['Day.1'].astype('category')
```

```
fig,ax=plt.subplots(figsize=(15,8))
plt.grid(True)
plt.title('High&Low Temperature per day')
plt.plot(df["Day"],df["High Temp (¬∞F)"],label='High Temp (∞F)')
plt.plot(df["Day"],df["Low Temp (¬∞F)"],label='Low Temp (∞F)')
plt.plot(df["Day"],df["Precipitation"],color='green')
plt.legend()
plt.xlabel('Date')
plt.show()
```

```
fig,ax=plt.subplots(figsize=(15,8))
sns.set(color_codes=True)
plt.title('Precipitation per day')
plt.plot(df["Day"],df["Precipitation"],color='green')
plt.xlabel('Date')
plt.ylabel('Precipitation')
```

```
sns.set(color_codes=True)
fig,ax=plt.subplots(figsize=(15,8))
sns.barplot(df['Day'],df['Total'])
```

```
sns.set(color_codes=True)
fig,ax=plt.subplots(figsize=(15,8))
plt.grid(True)
plt.title('Number of bicycles per bridge per day')
plt.plot(df["Day"],df['Brooklyn Bridge'],label='Brooklyn Bridge')
plt.plot(df["Day"],df["Manhattan Bridge"],label='Manhattan Bridge')
plt.plot(df["Day"],df['Williamsburg Bridge'],label='Williamsburg Bridge')
plt.plot(df["Day"],df["Queensboro Bridge"],label='Queensboro Bridge')
plt.legend()
plt.xlabel('Date')
plt.show()
```
